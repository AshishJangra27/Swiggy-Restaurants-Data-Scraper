```python
import json
import requests
import tqdm as tqdm
from bs4 import BeautifulSoup

res = requests.get('https://www.swiggy.com/#city-links')
soup = BeautifulSoup(res.content,'html.parser')
```

### 1. Scraping Cities Link

```python
cities = [[sp.text.capitalize(), 'https://www.swiggy.com' + sp.get('href')] for sp in soup.find_all('a', class_ = '_3TjLz b-Hy9')]
```

```python
len(cities)
```

### 2. Scraping Sub-Regions of each City - JSON

```python
dct = {}

for city,link in cities:           # Going through all the Cities
    
    soup = BeautifulSoup(requests.get(link).content, 'html.parser')   # Sending Request to Each City

    temp = {}                           # Dictionary to Save Subregions of each City
    
    if len(soup.find_all('div', class_ = '_1c_iw')):        # If Subregions are there
        for sp in soup.find_all('div', class_ = '_1c_iw')[0].find_all('a'):     
                                                            # Collecting the Subregions
            
            link_subregion = 'https://www.swiggy.com' + sp.get('href').strip()
            name_subregion = sp.text.strip()
 
            temp[name_subregion] = link_subregion           # Saving the data in a dictionary of each city        
            
        temp['link'] = link                                # Adding How many subregions are there
        dct[city] = temp                                    # Adding all subresions in the city dictionary
            
    else:                                                   # If Subregions are not there
        dct[city] =  link                                   # Adding link in the city dictionary
              
    print(city)
```

### 3. Saving the dataset

```python
fd = open('cities.json','w')
fd.write(json.dumps(dct))
fd.close()
```
